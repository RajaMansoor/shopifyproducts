```python
from selenium import webdriver
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.select import Select
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome('E:\Programming\chromedriver') 
```

```python
driver.get("https://www.shopistores.com/areas/9/")
driver.maximize_window()
```

```python
names = []
shop_urls = []
descriptions = []
fb_links = []
twitters = []
instagrams = []
printerests = []
snapchats = []
youtubes = []

for page in range(1, 25):
    print(page)
    page_url = f"https://www.shopistores.com/areas/9/{page}"
    driver.get(page_url)
    time.sleep(45)
    
    
    soup = BeautifulSoup(driver.page_source, "html.parser")
    trs = soup.find("table", {"id": "content-container-tbl"}).find("tbody").find_all("tr")

    for tr in trs:
        
        try:
            name = tr.find("td", {"data-title":"Store"}).contents[0].strip()
            names.append(name)
        except Exception as error:
            print(f"Error in name: {error}")
            names.append("")

        try:
            shop_url = tr.find("td", {"data-title":"Store"}).find("a").get("href")
            shop_urls.append(shop_url)
        except Exception as error:
            print(f"Error in shop url: {error}")
            shop_urls.append("")

        try:
            description = tr.find("td", {"data-title":"Store"}).find("small").get_text().strip()
            descriptions.append(description)
        except Exception as error:
            print(f"Error in description: {error}")
            descriptions.append("")

        try:
            fb_link = tr.find("td", {"data-title":"Social"}).find(lambda tag: tag.name == "a" and "facebook" in tag.get("href")).get("href")
            fb_links.append(fb_link)
        except Exception as error:
            print(f"Error in fb link: {error}")
            fb_links.append("")

        try:
            twitter = tr.find("td", {"data-title":"Social"}).find(lambda tag: tag.name == "a" and "twitter" in tag.get("href")).get("href")
            twitters.append(twitter)
        except Exception as error:
            print(f"Error in fb link: {error}")
            twitters.append("")

        try:
            instagram = tr.find("td", {"data-title":"Social"}).find(lambda tag: tag.name == "a" and "instagram" in tag.get("href")).get("href")
            instagrams.append(instagram)
        except Exception as error:
            print(f"Error in instagram link: {error}")
            instagrams.append("")

        try:
            printerest = tr.find("td", {"data-title":"Social"}).find(lambda tag: tag.name == "a" and "printerest" in tag.get("href")).get("href")
            printerests.append(printerest)
        except Exception as error:
            print(f"Error in printerest link: {error}")
            printerests.append("")

        try:
            snapchat = tr.find("td", {"data-title":"Social"}).find(lambda tag: tag.name == "a" and "snapchat" in tag.get("href")).get("href")
            snapchats.append(snapchat)
        except Exception as error:
            print(f"Error in snapchat link: {error}")
            snapchats.append("")

        try:
            youtube = tr.find("td", {"data-title":"Social"}).find(lambda tag: tag.name == "a" and "youtube" in tag.get("href")).get("href")
            youtubes.append(youtube)
        except Exception as error:
            print(f"Error in youtube link: {error}")
            youtubes.append("")
```

```python
import pandas as pd
```

```python
data = pd.DataFrame([names, shop_urls, fb_links, twitters, instagrams, printerests, snapchats, youtubes, descriptions])
data = data.transpose() 
data.columns=["Name", 'Shop_url', 'Facebook_Link', 'Twitter_Link', 'Instagram_Link', 'Printerest_Link', 'Snapchat_Link', 'Youtube_Link', 'Description']
data.head()
data.index = data.index + 1
```
