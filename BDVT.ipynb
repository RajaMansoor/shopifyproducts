```python
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
```

```python
profile_links =[]
first_names = []
last_names = []
phones = []
mobiles = []
faxes = []
webs = []
addresses = []

for page in range(1,35):
    
    
    source = requests.get(f'https://www.bdvt.de/bdvt/experten-expertinnen/?pageIdf324a716={str(page)}#list_f324a716')
    if source.status_code == 200:
        soup = BeautifulSoup(source.content, "html.parser")
        
        profiles = soup.find_all("li", {"class": "listEntry"})
        if profiles:
            print(f"Found Profiles: {str(len(profiles))}")
            
            for index, profile in enumerate(profiles):
                try:
                    first_name = " ".join(profile.find("div", {"class": "listEntryTitle"}).get_text().split(" ")[:-1])
                    print(first_name)
                    first_names.append(first_name)
                except Exception as error:
                    first_names.append("-")
                    print(f"Got Exception in first_name: {str(error)}")
                    
                try:
                    last_name = profile.find("div", {"class": "listEntryTitle"}).get_text().split(" ")[-1]
                    print(last_name)
                    last_names.append(last_name)
                except Exception as error:
                    last_names.append("-")
                    print(f"Got Exception in last_name: {str(error)}")
                    
                profile_link = f"https://www.bdvt.de{profile.get('data-url')}"
                if profile_link:
                    print(profile_link)
                    profile_links.append(profile_link)
                    
                    source = requests.get(profile_link)
                    time.sleep(0.6)
                    if source.status_code == 200:
                        soup = BeautifulSoup(source.content, "html.parser")
                        
                        try:
                            phone = soup.find("a", {"class": "phone"}).get_text().strip()
                            print(phone)
                            phones.append(phone)
                        except Exception as error:
                            phones.append("-")
                            print(f"Got Exception in phone: {str(error)}")
                            
                        try:
                            mobile = soup.find("a", {"class": "mobile"}).get_text().strip()
                            print(mobile)
                            mobiles.append(mobile)
                        except Exception as error:
                            mobiles.append("-")
                            print(f"Got Exception in mobile: {str(error)}")
                            
                        try:
                            fax = soup.find("a", {"class": "fax"}).get_text().strip()
                            print(fax)
                            faxes.append(fax)
                        except Exception as error:
                            faxes.append("-")
                            print(f"Got Exception in fax: {str(error)}")
                            
                        try:
                            web = soup.find("div", {"class": "elementSocialButtonType_varWebsite"}).find("a").get("href")
                            print(web)
                            webs.append(web)
                        except Exception as error:
                            webs.append("-")
                            print(f"Got Exception in web: {str(error)}")
                            
                        try:
                            address = soup.find("div", {"class": "listEntryContactPersonAddress"}).get_text().strip()
                            print(address)
                            addresses.append(address)
                        except Exception as error:
                            addresses.append("-")
                            print(f"Got Exception in address: {str(error)}")
                    
                    else:
                        profile_links.append("-")
                        print(f"Missed Profile: {profile_link}")
                    
                else:
                    print(f"Missed {str(index)} Profile From Page {str(page)}")
                
                
            
        else:
            print(f"Missed Profiles From: {str(page)}")
            
    else:
        print(f"Missed Page: {str(page)}")
    time.sleep(1)
```

```python
data = pd.DataFrame([first_names, last_names, phones, mobiles, faxes, profile_links, webs, addresses])
data = data.transpose() 
data.columns=['first_names', 'last_names', 'phones', 'mobiles', 'faxes', 'profile_links', 'webs', 'addresses']
data.head()
```

```python
data.to_csv('BDVT.csv', index=False)
```
